```python
import requests
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 100
seed = 556
min_works_count = 1000
r = requests.get(f"https://api.openalex.org/institutions?sample={n}&per-page={n}&seed={seed}&filter=works_count:>{min_works_count-1}")
```

```python
r.json()
```

```python
institutions = r.json()['results']
len(institutions)
```

```python
institutions[0]
```

```python
data = []
for institution in institutions:
    openalex_id = institution['id'].replace('https://openalex.org/', '')
    ror = institution['ror']
    openalex_count = institution['works_count']
    data.append({
        'openalex_id': openalex_id,
        'display_name': institution['display_name'],
        'ror': ror,
        'openalex_count': openalex_count,
    })
df_institutions = pd.DataFrame(data)
```

```python
df_institutions.to_csv('../data/institutions_minworks1000_random100.csv', index=False)
```
