```python
import requests
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 50
seed = 556
min_works_count = 10
r = requests.get(f"https://api.openalex.org/authors?sample={n}&per-page={n}&seed={seed}&filter=works_count:>{min_works_count-1}")
```

```python
r.json()
```

```python
authors = r.json()['results']
len(authors)
```

```python
authors[0]
```

```python
data = []
for au in authors:
    openalex_id = au['id'].replace('https://openalex.org/', '')
    display_name = au['display_name']
    h_index = au['summary_stats']['h_index']
    affil = au['last_known_institution'].get('display_name', None) if au['last_known_institution'] else None
    works_count = au['works_count']
    data.append({
        'openalex_author_id': openalex_id,
        'openalex_display_name': display_name,
        'openalex_num_papers': works_count,
        'openalex_affil': affil,
        'openalex_h_index': h_index,
    })

df_authors_alex_api = pd.DataFrame(data)
```

```python
df_authors_alex_api.to_csv('../data/random_openalex_authors_minworks10.csv', index=False)
```
