```python
# translated using DuckDuckGo
# from originals:
# ['Yayıncı',
#  'Başlık',
#  'Konu',
#  'Yazarlar',
#  'Anahtar Kelimeler',
#  'Doi',
#  'Veritabanı',
#  'Type',
#  'Atıf Sayısı',
#  'Cilt',
#  'Sayı',
#  'Issn',
#  'e-issn',
#  'Yayın Tarihi',
#  'Dspace',
#  'Link']
columns_english = ['Publisher', 'Title', 'Subject', 'Authors', 'Keywords', 'doi', 'Database', 'Type', 'Number of Citations', 'Skin', 'Number', 'Issn', 'e-issn', 'Release Date', 'Dspace', 'Link']
```

```python
df = pd.read_csv('../data/Bartın Üniversitesi.csv', names=columns_english, header=0)
```

```python
df.columns.tolist()
```

```python
email = "jportenoy@ourresearch.org"

dois = df['doi'].dropna().unique()

endpoint = "works"
size = 50
loop_index = 0
works = []
for list_index in range(0, len(dois), size):
    subset = dois[list_index:list_index+size]
    pipe_separated_ids = "|".join(subset)
    r = requests.get(f"https://api.openalex.org/{endpoint}?filter=doi:{pipe_separated_ids}&per-page={size}&mailto={email}")
    results = r.json()['results']
    works.extend(results)
    loop_index += 1
print(f"collected {len(works)} {endpoint} using {loop_index} api calls")
```

```python
df['doi'].duplicated().value_counts()
```

```python
works[0]['authorships']
```

```python
data = []
for work in works:
    this_institutions = []
    for authorship in work['authorships']:
        for institution in authorship['institutions']:
            if 'id' in institution:
                this_institutions.append(institution['id'])
    data.append({
        'openalex_id': work['id'],
        'doi': work['doi'].replace('https://doi.org/', ''),
        'institution_match': 'https://openalex.org/I254571519' in this_institutions,
    })
df_openalex_query = pd.DataFrame(data)
```
